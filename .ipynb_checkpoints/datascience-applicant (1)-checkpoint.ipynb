Since the longitude and latitude values are floats, we can continue.

```python
'''Distance equation for long,lat data used via stackoverflow from user Michael0x2a. 
Updated to a function that converts to mileage'''
# constant values, if need to change end lat, long points, change the lat2, lon2 information
lat2 = np.array(clean.Latitude)
lon2 = np.array(clean.Longitude)
latr = np.array(list(map(lambda x: np.radians(x), lat2)))
lonr = np.array(list(map(lambda x: np.radians(x), lon2)))
def distance(lat1,lon1):
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    dlon = np.array(list(map(lambda x: (x - lon1), lonr)))
    dlat = np.array(list(map(lambda x: (x - lat1), latr)))
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    # 6373.0 represents earth radius in kilometers
    kilo = 6373.0 * c
    miles = kilo * 0.62137119
    return miles
```

## Model building

Train a statistical model of your choosing to perform this prediction task. Describe each step of your code in comments and explain your rationale for choosing a particular model.

Note: We do not expect incredibly high model accuracy. Nevertheless, if model accuracy is low, please include a written discussion to explain why this might be the case. We want to understand your thought process.

For this project, we will utilize a time-series model to predict future crime incidents. Some initial analysis will be performed before predicting. The incident number will be used to identify each instance. The data will be decomposed to check for trends, seasonality and residuals. Those areas will be analyzed before putting the data through a model that will predict with a certain number of future days.

Facebook Prophet has been shown to be a reliable forecasting model. This is the model that we will utilize for this time-series forecasting.

```python
#read in series for time-series analysis
ts = pd.read_csv('engine_Austin.csv', usecols=['Incident Number','Occurred Date Time'])
```

```python
ts.head()
```

```python
# manipulating data to run decomposiition
ts['Occurred Date Time'] = pd.to_datetime(ts['Occurred Date Time'])
ts.set_index('Occurred Date Time', inplace=True)
```

```python
ts.info()
```

```python
ts.index.year.value_counts()
```

```python
ts1 = ts.resample('D').count()
```

```python
ts1.head()
```

```python
ts1['Incident Number'].mean()
```

Since I had this data available, I wanted to look into the average number of incidents per day and add it to insights.

```python
#decompose the data to check for trend, seasonality and noise (residuals)
decomposition = seasonal_decompose(np.log(ts1),model='additive')
# Gather the trend, seasonality and noise of decomposed object
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot gathered statistics
plt.figure(figsize=(12,8))
plt.subplot(411)
plt.plot(np.log(ts), label='Original', color="blue")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend', color="blue")
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality', color="blue")
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals', color="blue")
plt.legend(loc='best')
plt.tight_layout()
```

Based on the decomposition of daily incident counts, there isn't a trend that can be followed until the end of the year, where incidents drop significantly. There is a definite seasonality to the data and there is adequate noise from residuals. 

```python
#autocorrelation visual
plt.figure(figsize=(12,6))
pd.plotting.autocorrelation_plot(ts1)
```

Autocorrelation shows a high significance at the first few lags, with random peaks at 75, 160 and 250. 

```python
#partial autocorrelation visual
rcParams['figure.figsize'] = 14, 5
plot_pacf(ts1, lags = 100);
```

The partial autocorrelation visual shows that there is significance in the first couple of lags. Interstingly, there is significance around the early 60s and late 80s lags as well.

```python
#manipulating data to run in Facebook Prophet
ts1.index.rename('ds', inplace = True)
ts1.rename(columns = {'Incident Number':'y'}, inplace = True)
ts1.reset_index(level=0, inplace=True)
```

```python
ts1.head()
```

```python
# set the uncertainty interval to 95%, remove weekly seasonality
Model = proph(interval_width=0.95, weekly_seasonality = False)
# add daily seasonality adjustment
Model.add_seasonality(name='daily', period=1, fourier_order = 3)
# Fit the timeseries into Model
train = Model.fit(ts1)
# Use make_future_dataframe with a monthly frequency and periods - 1 month
future_dates = Model.make_future_dataframe(periods=1, freq='M')
future_dates.tail()
#This future dates dataframe can now be used as input to the predict method of the fitted model.   
# Predict the values for future dates and take the head of forecast
forecast = Model.predict(future_dates)
forecast.head()
```

```python
Model.plot(forecast, uncertainty=True) 
plt.show()
```

```python
Model.plot_components(forecast) 
plt.show()
```

Based on the forecasted model, crime incidents will continue to downtrend in the month following the end of 2018.
